```python
data['rate'] = data['rate'].replace('/5','')
data['rate'] = data['rate'].astype(float)
data.info()
```

```python
binary_cols = ['online_order','book_table']
for col in binary_cols:
  data[col] = data[col].map({'Yes':1,'No': 0})
```

```python
data
```

```python
data['votes'] = data['votes'].replace(0,data['votes'].median())
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].replace(0,data['approx_cost(for two people)'].median())
```

```python
data
```

```python
data = data.drop('name',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
LB = LabelEncoder()
data['listed_in(type)'] = LB.fit_transform(data['listed_in(type)'])
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
data[['votes' , 'approx_cost(for two people)']] = Scaler.fit_transform(data[['votes' , 'approx_cost(for two people)']])
```

```python
data
```

```python
data[['votes' , 'approx_cost(for two people)']] = data[['votes' , 'approx_cost(for two people)']].round(3)
```
